```python
import pandas as pd
import re
import matplotlib.pyplot as plt
from IPython.core.display import display, HTML
from datetime import datetime, timedelta

import warnings
warnings.filterwarnings("ignore")

def data_from_url(url):
    df = pd.read_html(url)[1]
    lol = df.to_numpy().tolist()
    return lol

def fetch_year(date_string):
    return int(re.findall("\d{4}", date_string)[0])
```

```python
def find_dates(date_string):
    return int(re.findall("\d{4}", date_string)[0])
```

```python
def barplot(list_of_2_element_list):
    d = {ya[0]: ya[1] for ya in list_of_2_element_list} ### dictionary comprehensions
    plt.figure(figsize=(9,15))
    
    axes = plt.axes()
    axes.get_xaxis().set_visible(False)
    spines = axes.spines
    spines['top'].set_visible(False)
    spines['right'].set_visible(False)
    spines['bottom'].set_visible(False)
    spines['left'].set_visible(False)
    
    ax = plt.barh(*zip(*d.items()), height=0.5) ## what does this mean
    plt.yticks(list(d.keys()), list(d.keys()))
    plt.xticks(range(4), range(4))
    
    rectangles = ax.patches
    
    for rectangle in rectangles:
        x_value = rectangle.get_width()
        y_value = rectangle.get_y() + rectangle.get_height() / 2
        space = 5
        ha = 'left'
        label = "{}".format(x_value)
        if x_value  > 0:
            plt.annotate(
                label,
                (x_value, y_value),
                xytext = (space, 0),
                textcoords = "offset points",
                va = "center",
                ha= ha
            )
            
        axes.tick_params(tick1On = False)
        plt.show()
        
    
```

```python
def unique_countrues(countries):
    s = pd.Series(countries)
    return list(s.unique())

def display_no_index(df):
    display(HTML(df.to_html(index=False)))
    
def print_pretty_table(countries_frequency):
    countries = df.Country.value_counts().index
    occurences = df.Country.value_counts().values
    d = {"Country": countries, "Number of Occurences": occurences}
    display_no_index(pd.DataFrame(d))
```

```python
df = pd.read_html("https://en.wikipedia.org/wiki/List_of_helicopter_prison_escapes")[1]
df = df[["Date","Prison name","Country","Succeeded","Escapee(s)"]]
```

```python
df.Country.value_counts()
```

```python
url = "https://en.wikipedia.org/wiki/List_of_helicopter_prison_escapes"
data = data_from_url(url)
```

```python
for row in data:
    row[0] = fetch_year(row[0])
```

```python
min_year = min(data, key=lambda x: x[0])[0]
max_year = max(data, key = lambda x: x[0])[0]

years = []
for year in range(min_year, max_year + 1):
    years.append(year)
    
attempts_per_year = []
for year in years:
    new_data = [year,0]
    attempts_per_year.append(new_data)
    
print(attempts_per_year[:3])    
    
for row in data:
    for year_attempt in attempts_per_year:
        if row[0] == year_attempt[0]:
            year_attempt[1] += 1
            
print(attempts_per_year)                

%matplotlib inline
barplot(attempts_per_year)
```

```python
dates = df["Date"].to_numpy().tolist()

formatted_date = []

for date in dates:
    try:
        formatted_date.append(datetime.strptime(date, "%B %d, %Y"))
    except ValueError:
        try:
            formatted_date.append(datetime.strptime(date, "%B, %Y"))
        except ValueError: 
            formatted_date.append(datetime.strptime(date, "%B %Y"))
df["Formatted_date"] = formatted_date  
```
